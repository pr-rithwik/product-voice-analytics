```python
from huggingface_hub import login, upload_file
from google.colab import userdata
from src.config import (
    TFIDF_VECTORIZER_PATH,
    LR_MODEL_PATH,
    DISTILBERT_PATH
)

login(token=userdata.get('HF_TOKEN'))
```

```python
from huggingface_hub import HfApi

api = HfApi()
repo_id = 'rithweek/product-voice-analytics-models'

# upload tfidf vectorizer
api.upload_file(
    path_or_fileobj=TFIDF_VECTORIZER_PATH,
    path_in_repo='tfidf_vectorizer.pkl',
    repo_id=repo_id,
    repo_type='model'
)

# upload lr model
api.upload_file(
    path_or_fileobj=LR_MODEL_PATH,
    path_in_repo='lr_model.pkl',
    repo_id=repo_id,
    repo_type='model'
)

# upload distilbert folder
api.upload_folder(
    folder_path=DISTILBERT_PATH,
    path_in_repo='distilbert/',
    repo_id=repo_id,
    repo_type='model'
)
```
